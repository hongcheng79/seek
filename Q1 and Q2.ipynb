# Question 1

```python
def question_1(array):
    input = np.array(array)
    result = np.array([])
    
    # start from index 1, so the minus is valid for first lower
    for index in range(1, input.size+1):
        # lower limit + upper limit
        lower = np.sum(input[0:index-1])
        upper = np.sum(input[index:input.size])
        result = np.append(result,lower+upper)
        
    return result

class Question1UnitTest(unittest.TestCase):
    def test_run(self):
        input = question_1([1,2,3,4])
        expected = np.array([9,8,7,6])
        self.assertTrue(np.array_equal(input,expected))
        
suite = unittest.TestLoader().loadTestsFromModule(Question1UnitTest())
unittest.TextTestRunner().run(suite)
```

# Question 2

```python
# Download data
question2_data_path = data_path + '/weekly-sales-transactions'
question2_data_file = question2_data_path +'/Sales_Transactions_Dataset_Weekly.csv'

if not os.path.exists(question2_data_path):
    os.mkdir(question2_data_path)

if not os.path.exists(question2_data_file):
    print('Downloading data as '+question2_data_file)
    os.system('cd '+question2_data_path+';kg dataset -o crawford -d weekly-sales-transactions')
else:
    print('Data file exist '+question2_data_file)
    
# Load data up
question2_data = pd.read_csv(question2_data_file)
question2_data.head()
```

```python
# Split the data into 2 panda frames - one hold quantity, one hold normalized
question2_data_quantity = pd.DataFrame()
question2_data_normalized = pd.DataFrame()

for index in range(0, 50):
    question2_data_quantity['W'+str(index)] = question2_data['W'+str(index)]
    question2_data_normalized['Normalized '+str(index)] = question2_data['Normalized '+str(index)]
    
question2_data_quantity.head()
```

e) Identify the top 5 worst performing products on a biweekly basis

```python
column_size = question2_data_normalized.shape[1]

for index in range(0, column_size, 2):
    # plus n with n + 1 and divide by 2
    section_e = (question2_data_normalized['Normalized '+str(index)] 
                 + question2_data_normalized['Normalized '+str(index+1)]) / 2
    # sort by ascending order (near 0.0) and get the first 5
    indexes = section_e.sort_values().head().index.tolist()
    print('W'+str(index)+':'+str(question2_data.loc[indexes]['Product_Code'].values))
    
```

f) Identify outliers from the data and output the corresponding week numbers

```python
# using Z-score method from http://colingorrie.github.io/outlier-detection.html
def outliers_z_score(ys):
    threshold = 3

    mean_y = np.mean(ys)
    stdev_y = np.std(ys)
    z_scores = [(y - mean_y) / stdev_y for y in ys]
    return np.where(np.abs(z_scores) > threshold)

# Loop through the product list
for index, row in question2_data_quantity.iterrows():
    outliers = outliers_z_score(row)
    if outliers[0].size > 0:
        week = ['W{0}'.format(i) for i in outliers[0]]
        print(question2_data.loc[index]['Product_Code']+':'+str(week))
```
